**3) What are pre-defined keywords in Python ?**

**5) What is mutability in Python ?**

**9) What is type casting in Python ?**

      Typecasting in python is just changing the data type of one variable to another data type.

**10) What is the difference between implicit and explicit type casting ?**

       - Implicit typecasting means : automatic conversion of the datatype.

**13) What is the difference between for and while loops ?**

```python
# 10) Write code to demonstrate type casting with list elements.
# List containing strings of numbers
str_list = ["10", "20", "30", "40"]

# Explicit type casting: converting each element to int
int_list = [int(x) for x in str_list]

print("Original list (string):", str_list)
print("After type casting to int:", int_list)


# Another example: converting int list to float
float_list = [float(x) for x in int_list]
print("After type casting to float:", float_list)


# Mixed-type casting example
mixed = ["5", 8, "12.5", 3.4]
converted = [float(i) for i in mixed]   # converting all to float
print("Mixed list:", mixed)
print("After type casting to float:", converted)
```

```python
# 13) Write a Python program to find the sum of all even numbers between 1 and 50.
total = 0

for num in range(1, 51):   # range(1,51) means 1 to 50
    if num % 2 == 0:        # check even
        total += num

print("Sum of even numbers between 1 and 50 is:", total)

```

```python
# 14) Write a program to reverse a string using a while loop.
# Program to reverse a string using while loop

text = input("Enter a string: ")

reverse_str = ""
i = len(text) - 1          # Start from last index

while i >= 0:
    reverse_str += text[i]
    i -= 1

print("Reversed string:", reverse_str)
```

```python
# 15) Write a Python program to calculate the factorial of a number provided by the user using a while loop.
# Program to find factorial of a number using while loop

num = int(input("Enter a number: "))

factorial = 1
i = 1

while i <= num:
    factorial *= i
    i += 1

print("Factorial of", num, "is:", factorial)
```
